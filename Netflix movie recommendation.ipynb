# Demographic Filtering

-- to find the rating we are using IMDBs (Internet Movie Database) weighted rating 
-- The idea of weighted rating is to provide more meaningful rankings by giving more weight to movies with higher average votings and a significant number of votes, while avoiding unfair rankings due to movies with very few votes.
-- Its a common approach to create a more reliable and robust ranking system in various movie recommendation scenarios.
-- weighted rating=(v/(v+m)*R)+(m/(v+m)*C)
v is the number of votes for the movie;
m is the minimum votes required to be listed in the chart;
R is the average rating of the movie; And
C is the mean vote across the whole report

```python
#to find the C
C=df['vote_average'].mean()
C
#So, on a scale of 1 to 10, the average rating for all the movies is approximately 6.
```

```python
#to determine the value of "m," the minimum number of votes necessary for a movie to appear in the chart,
#we will set the cutoff at the 90th percentile. 
#This means that a movie must receive more votes than at least 90% of the other movies on the list to be included in the ranking or chart.
m= df['vote_count'].quantile(0.9)
m
```

```python
#filtering the movies based on this m value
df_qm = df.copy().loc[df['vote_count'] >= m]
df_qm.shape
```

```python
#481 movies have been qualified 
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
df_qm['rate'] = df_qm.apply(weighted_rating, axis=1)

df_qm.head(5)
```

```python
#sorting the movies based on rate 
df_qm = df_qm.sort_values('rate', ascending=False)

#Printing the top 10 movies
df_qm[['title_x', 'vote_count', 'vote_average', 'rate']].head(10)
```

# Content Based Filtering

In this recommender system, the movie content, including details such as the movie overview, cast, crew,genre,tagline etc., is utilized to identify similarities between movies.
By analyzing these content attributes, the system identifies movies that share common characteristics and themes. 
Based on these similarities, the system recommends movies that are most likely to be similar to the ones the user has shown interest in.

Description based recommender 
#plot description-based recommender system operates by computing pairwise similarity rates among all movies using their descriptions. 
#This approach allows the system to identify movies with similar plot themes and content. 

```python
#here description is in the overview column
df['overview'].head()
```

```python
#using TF-IDF to convert the textual plot descriptions into numerical feature vectors. 
#TF-IDF is a common text vectorization technique that allows  to represent the importance of each word in a document relative to the entire document.
from sklearn.feature_extraction.text import TfidfVectorizer
#removing all english stop words
Tfidf=TfidfVectorizer(stop_words='english')
df['overview']=df['overview'].fillna('')
Tfidf_matrix=Tfidf.fit_transform(df['overview'])
Tfidf_matrix.shape
```

```python
#20,978 different words were used to describe the 4803 movies in our dataset.
```

```python
#using cosine similarity to find the similarity score. 
#Since we have used the TF-IDF vectorizer, calculating the dot product will directly give us the cosine similarity score.
#Therefore, we will use sklearn's linear_kernel() instead of cosine_similarities() since it is faster.
```

```python
from sklearn.metrics.pairwise import linear_kernel

# Computing the cosine similarity matrix
cosine_sim = linear_kernel(Tfidf_matrix, Tfidf_matrix)

cosine_sim
```

```python
#To create a function that accepts a movie title as input and returns a list of the ten most comparable films, 
#we need to first establish a reverse mapping between the DataFrame indices and movie titles. 
#This reverse mapping will enable us to recognize the index of a movie given its title, allowing us to efficiently access
#the relevant information in the metadata DataFrame
```

```python
#Constructing a reverse map of indices and movie titles
indices = pd.Series(df.index, index=df['title_x']).drop_duplicates()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Function that takes in title as input and outputs wil be the most similar movies
def get_recommendations(title_x, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title_x
    idx = indices[title_x]
    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['title_x'].iloc[movie_indices]
```

#We will use these functions to preprocess the metadata and create new columns in the DataFrame with the extracted information

```python
#The ast module in Python stands for Abstract Syntax Trees. It provides a way to work with and manipulate Python abstract syntax trees, which represent the syntactic structure of Python code without including any details about the actual execution.
#so here we are using 'literal_eval' library from the ast module parse the stringified code into python objects (like list,tuple,dict etc).so for that we are taking features such as 
#'cast','crew','keywords','spoken_languages','genres'
```

```python
#we need to convert the features that are currently represented as strings (stringified) in the data into their appropriate Python data structures. 
```

```python
features= df[['genres','keywords','cast','crew','spoken_languages']]
```

```python
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres','spoken_languages']
for feature in features:
    df[feature] = df[feature].apply(literal_eval)
```

```python
## Getting the director's name from the crew feature. If director is not listed, return NaN
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# Returning the list top 3 elements or entire list; whichever is more.
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Checking if more than 3 elements exist. If yes, returns only first three. If no, returns entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
df['director'] = df['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres','spoken_languages']
for feature in features:
    df[feature] = df[feature].apply(get_list)
```

```python
df[['title_x', 'cast', 'director', 'keywords', 'genres','spoken_languages']].head()
```

```python
#The next step is to convert the keywords and occurrences to lowercase and remove any spaces between them so that our vectorizer does not count 'Tom Hanks'
#'Tom Hiddles' as Tom.
```

```python
# defining the Function to convert all strings to lower case and avoid spaces btw names
def modified_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres','spoken_languages']

for i in features:
    df[i] = df[i].apply(modified_data)
```

```python
#now we can concat the string containing all the relevant metadata information that we want to include in a vectorization process.
#In this context, the metadata refers to specific details about the movie, such as the actors, director, and keywords associated with it.
```

```python
def concat(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres']) + ' ' + ' '.join(x['spoken_languages'])
df['soup'] = df.apply(concat, axis=1)
```

```python
#In the descriprion based filtering approach we have used TF-IDF,here  we use the CountVectorizer() instead of TF-IDF to avoid underestimating the importance of actors/directors 
#who have been involved in more movies.
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

# Collaborative Filtering 

```python
#Both scalability and sparsity pose significant challenges for collaborative filtering algorithms
#as they can result in increased computational complexity and reduced prediction accuracy. 
#Latent factor models(which is a type of CF) like Singular Value Decomposition (SVD) and Matrix Factorization are used to address these challenges 
#and improve the efficiency and accuracy of collaborative filtering-based recommendation systems.
```

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection import cross_validate
from surprise.model_selection import KFold
```

```python
reader = Reader()
ratings = pd.read_csv('ratings_small.csv')
ratings.head()
```

```python
data= Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
kf = KFold(n_splits=5)
```

```python
svd = SVD()
# Run 5-fold cross-validation and then print results
cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
#here we got a test RMSE of 0.8940 which is a good value
```

```python
trainset = data.build_full_trainset()
svd.fit(trainset)
```

# Conclusion

-- I have created recommenders using demographic , content- based and collaborative filtering. While demographic filtering is very elemantary and cannot be used practically, Hybrid Systems can take advantage of content-based and collaborative filtering as the two approaches are proved to be almost complimentary.

-- In content-based filtering, the recommendation system utilizes features such as 'genres', 'keywords', 'cast', 'crew', and 'spoken_languages' of movies to suggest similar movies to users. By analyzing these movie attributes, the system identifies patterns and similarities between movies, allowing it to recommend films that share common characteristics with those the user has shown interest in.

-- In general For new users, content-based filtering analyzes the features of items they have shown interest in, such as movie genres, actors, directors, keywords, or any other relevant attributes. It then uses this information to find similar items in the database that share similar attributes. By recommending items that align with the user's expressed interests, content-based filtering can provide a starting point for the new user to explore and discover content that suits their preferences.

-- On the other hand, in collaborative filtering I have used Singular Value Decomposition (SVD) as a technique to make recommendations. SVD is a matrix factorization method that uncovers latent factors representing underlying user preferences and item attributes. By breaking down the user-item interaction matrix, SVD identifies latent features that influence user-item interactions. This enables the system to recommend movies based on the preferences of similar users, as it finds users with similar tastes and suggests movies that have been positively rated by those users.

-- The cold start problem is a common challenge in recommendation systems that occurs when there is insufficient or no data available for a new user or item.in this situation by combining both collaborative filtering and content-based filtering techniques in a hybrid approach will be effective to overcome this problem.
